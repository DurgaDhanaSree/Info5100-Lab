```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('bank_transactions_data_2.csv')
```

```python
dataframe
```

```python
dataframe.head(-10)
```

```python
# Line Plot Univariate
plt.plot(dataframe["TransactionID"], dataframe['TransactionAmount'] )
plt.title("Bank Transactions")
plt.xlabel("TransactionID")
plt.ylabel("Transactions Amount")
plt.show()
```

```python
#plt.grid(True)

# Scatter Plot
plt.scatter(dataframe["TransactionID"], dataframe["TransactionAmount"])
plt.title("Transactions")
plt.xlabel("IDs")
plt.ylabel("Amount")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataframe['TransactionAmount'], dataframe['AccountBalance'], alpha=0.6)
plt.title('Scatter Plot: Account Balance vs Transaction Amount', fontsize=14)
plt.xlabel('Transaction Amount', fontsize=12)
plt.ylabel('Account Balance', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(dataframe['TransactionAmount'], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Transaction Amount', fontsize=14)
plt.xlabel('Transaction Amount', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(dataframe['CustomerAge'], bins=20, color='green', alpha=0.7, edgecolor='black')
plt.title('Histogram of Customer Age', fontsize=14)
plt.xlabel('Customer Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Box plot for TransactionAmount
plt.figure(figsize=(10, 6))
sns.boxplot(x=dataframe['TransactionAmount'], color='skyblue')
plt.title('Box Plot of Transaction Amount', fontsize=14)
plt.xlabel('Transaction Amount', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
